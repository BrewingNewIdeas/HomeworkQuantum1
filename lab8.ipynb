```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi
from mpl_toolkits.mplot3d import Axes3D
```

```python
def S(E1,E2,E3,E4):
    return E1-E2+E3+E4
```

```python
def Ephi(the,thet):
    return np.cos(2*(the-thet))
```

```python
def Emix(thet1,thet2):
    return np.cos(2*thet1)*np.cos(2*thet2)
```

```python
thet1=0
thet1p=pi/4
thet2=np.linspace(0,2*pi,1000)
thet2p=np.linspace(0,2*pi,1000)

X,Y=np.meshgrid(thet2,thet2p)
Z=S(Emix(thet1,X),Emix(thet1,Y),Emix(thet1p,X),Emix(thet1p,Y))
np.shape(X),np.shape(Y),np.shape(Z)
points=Z>round(2*np.sqrt(2),4)





fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
cp = ax.contour3D(X, Y, Z, 1000)  # Increase the number of contours for smoother transitions

fig.colorbar(cp)  # Adds a color bar to show the contour levels
ax.set_title('3D Contour Map')
ax.set_xlabel('X coordinate')
ax.set_ylabel('Y coordinate')
ax.set_zlabel('Z value')
plt.show()


```
